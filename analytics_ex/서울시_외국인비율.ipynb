```python
import pandas as pd

df = pd.read_excel("datas/인구.xlsx")

# 1) '동별(1)' 컬럼 위에서 아래로 채우기 (NaN인 곳에 바로 위 값 복사)
df["동별(1)"] = df["동별(1)"].replace("동별(1)", pd.NA)  # 맨 위 제목 셀 정리용
df["동별(1)"] = df["동별(1)"].ffill()

# 2) '등록외국인' 행만 필터 + '합계'(서울 전체) 행은 제외
foreigners = df[(df["항목"] == "등록외국인") & (df["동별(1)"] != "합계")]

# 3) 구 이름 + 외국인 합계 컬럼만 추출
result = foreigners[["동별(1)", "2025 2/4"]].copy()
result.columns = ["구", "외국인수"]

# 4) 숫자로 변환 (콤마 제거)
result["외국인수"] = (
    result["외국인수"].astype(str).str.replace(",", "", regex=True).astype(int)
)

print(result)

```
