```python
import requests
import folium
# from pandas.io.json import json_normalize <-- 아래와 같이 수정
from pandas import json_normalize
```

```python
# 스벅 위치 데이터 가져오기
targetSite='https://www.starbucks.co.kr/store/getStore.do?r=8K3XF06R9Q'
req=requests.post(targetSite, data={
'ins_lat': 37.563398,    # 위도
'ins_lng': 126.9863309,  # 경도
'p_sido_cd': '01',
'p_gugun_cd': '',
'in_biz_cd': '',
'iend': 300,   # 300개
'set_date': ''
})
```

```python
starbucks=req.json()
```

```python
print(type(starbucks))
print(starbucks.keys())
```

```python
data= {
    "status": "success",
    "count": 5,
    "list": [
        {"store_id": "001", "name": "Store A", "location": "Seoul"},
        {"store_id": "002", "name": "Store B", "location": "Busan"}
    ]
}
data
```

```python
starbucks_df=json_normalize(starbucks, 'list')
starbucks_df
```

```python
# df 정보 보기
# starbucks_df.info()
starbucks_df.info(verbose=True, memory_usage='deep', show_counts=True)
```

```python
starbucks_df.columns
```

```python
cols = ['s_name','sido_code','sido_name','gugun_code','gugun_name', 'doro_address','lat','lot']
starbucks_df_map=starbucks_df[cols]
starbucks_df_map.head(3)
```

```python
starbucks_df_map.info()
```

```python
# 위도, 경도 숫자로 변경
starbucks_df_map.loc[:, 'lat']=starbucks_df_map['lat'].astype(float)
starbucks_df_map.loc[:, 'lot']=starbucks_df_map['lot'].astype(float)
```

```python
# 지도에 표현하기
starbucksMap=folium.Map(location=[37.5700249, 126.9836398], zoom_start=16)
popup=folium.Popup('종각점', max_width=300)
folium.Marker(location=[37.569918, 126.984528], popup=popup).add_to(starbucksMap)
popup=folium.Popup('종로관철점', max_width=300)
folium.Marker(location=[37.569058, 126.986013], popup=popup).add_to(starbucksMap)
popup=folium.Popup('종로구청점', max_width=300)
folium.Marker(location=[37.572662, 126.981138], popup=popup).add_to(starbucksMap)
starbucksMap.save('map_data/starMap1.html')
starbucksMap
```

## [문제] 위에있는 모든 지역을 지도에 표현하기

```python
starbucks_df_map.head(3)
```

```python
starbucks_map2 = starbucks_df_map.loc[:,['s_name','lat','lot']]
starbucks_map2.head()
```

```python
s_map_datas = starbucks_map2.values
s_map_datas[:5]
```

```python
# 지도 객체 생성
foliumMap = folium.Map(location=[s_map_datas[0][1], s_map_datas[0][2]], zoom_start=16)

# map에 표시하기 (모든 매장 마커 추가)
for name, lat, lot in s_map_datas:
    folium.Marker(
        [lat, lot],       # 위도, 경도
        popup=name        # 매장 이름
    ).add_to(foliumMap)

# 지도 파일로 저장
foliumMap.save('map_data/folium_map1.html')

# foliumMap 객체 반환
foliumMap
```
